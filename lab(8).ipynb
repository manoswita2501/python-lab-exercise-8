```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the dataset
df = pd.read_csv('Titanic - Titanic.csv')

# Step 2: Display the first and last 10 instances from the dataset
print("First 10 instances:")
print(df.head(10))
print("\nLast 10 instances:")
print(df.tail(10))
```

```python
# Step 3: Acquire necessary information
print("\nInformation about the dataset:")
df.info()
print("\nDescriptive statistics:")
print(df.describe())
```

```python
# Step 4: Retrieve the number of columns and rows (using shape)
num_rows, num_cols = df.shape
print(f"\nNumber of rows: {num_rows}")
print(f"Number of columns: {num_cols}")
```

```python
# Step 5: Data Visualization using Matplotlib
# Visualize the Gender of Passengers using a Bar graph
gender_counts = df['Sex'].value_counts()
plt.figure(figsize=(8, 4))
plt.bar(gender_counts.index, gender_counts.values)
plt.title("Gender of Passengers")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Visualize the Survival Count of Passengers using a Bar graph
survival_counts = df['Survived'].value_counts()
plt.figure(figsize=(8, 4))
plt.bar(survival_counts.index, survival_counts.values)
plt.title("Survival Count of Passengers")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# Visualize the Age of Passengers using a Histogram graph
plt.figure(figsize=(8, 4))
plt.hist(df['Age'], bins=20, edgecolor='k')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Visualize the comparison of Age and Fare of Passengers using a Scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.title("Age vs. Fare of Passengers")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
